## Iterating over sequences

````python
numbers = [1, 2, 3]

for n in numbers:
  print(n**2)
````

We can perform the same operation using a list comprehension, which takes the following form:

```python
[item for item in iterable_sequence]
```

````python
[n**2 for n in numbers]
````

A list comprehension can also have an if statement that guards the elements entering the iteration

````python
[n for n in numbers if n >= 2]
````

We can use a custom function, usually written as a lambda function to change the order of sorting with a key. In the following list, each element is another list that contains the name of the item in element `0` and the amount of stock in element `1`:

````python
stock = [
    ['apples', 10],
    ['pears', 2],
    ['oranges', 5],
]
````

The `sorted()` function automatically sorts by the first item in a list, so it would return an alphabetically sorted list: 

````python
sorted(stock)
````

If we to sort by the second element in the list, we have to write a function that will return the first element from each list that the `sorted()` function can use as a key for sorting

In the example below, the lambda function takes an object, and returns the value in index `1`. Which in our case the the number of items we have in stock.

````python
sorted(stock, key=lambda x: x[1])
````

We can use the `filter()` function to filter items inside a list using a similar lambda function. Here I filter for items that are more than five in stock:

````python
for item in filter(lambda x: x[1] > 5, stock):
    print(item)
````

We can achive the same using a list comprehension:

````python
[x for x in stock if x[1] > 5]
````
